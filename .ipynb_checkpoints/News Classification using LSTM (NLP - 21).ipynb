### Data Cleaning

```python
# The next step is data cleaning.
# So first we collect all those publishers from 'text' column of real dataset which are unknown and the news is not came 
# from twitter.
unknown_publishers = []
for index, row in enumerate(real.text.values):
    try:
        record = row.split('-', maxsplit=1)
        record[1]
        
        assert(len(record[0])<120)
    except:
        unknown_publishers.append(index)
```

```python
# So to see how many unknown publishers are there:
len(unknown_publishers)  
```

```python
# Now to see the text of the unknown publishers we use iloc function:
real.text.iloc[unknown_publishers]
```

```python
# So here we have an empty row and we need to remove it:
real = real.drop(8970, axis = 0)
```

```python
# Next we take the publishers and text from the text column of real news dataset:
publishers = []
temp_text = []

for index, row in enumerate(real.text.values):
    if index in unknown_publishers:
        temp_text.append(row)
        publishers.append("Unknown")
        
    else:
        record = row.split('-', maxsplit = 1)
        publishers.append(record[0].strip())
        temp_text.append(record[1].strip())
```

```python
# So next we replace the text column of the real data with the 'temp_text' which we extracted and also we create a new 
# column 'publisher' to store the publisher of the relavant text.
real['publisher'] = publishers
real['text'] = temp_text
```

```python
# Now to see few of the samples:
real.sample(5)
```

```python
# The shape of the dataset is:
real.shape
```

```python
# Next let's see the fake news if we have empty rows overthere.
# So for that we do list comprehension:
empty_fake_index = [index for index, text in enumerate(fake.text.to_list()) if str(text).strip()==""]
len(empty_fake_index)
```

```python
# So we remove these empty rows. 
# To see the empty rows:
fake.iloc[empty_fake_index]
```

```python
# So as the 'text' column text is present in the 'title' column, we merge both the columns in both datasets:
real['text'] = real['title'] + " " + real['text']
fake['text'] = fake['title'] + " " + fake['text']
```

```python
# Next we convert the text of text column from lowercase into uppercase:
real['text'] = real['text'].apply(lambda x: str(x).lower())
fake['text'] = fake['text'].apply(lambda x: str(x).lower())
```

## Preprocessing text

```python
# So next we need to convert the text data into numeric data, because ML don't understand text. So for that here we use 
# 'word2Vec' technique. It will convert each of word into a sequence of vectors. So thats what we're going to do now. 
# This process is normally called word embedding.
# So first we import the gensim library and we put the target input into the 'y' variable which is the class labels.
import gensim
y = dataset['class'].values
```

```python
# Next we take the text data 'column text' in a form of list data.
X = [d.split() for d in dataset["text"].to_list()]  # it will represent the data as sequence of list.
```

```python
print(X[0])
```

```python
# Now the X will be feed into the gensim model word2vec conversion.
# So the vector size we will use here is 100. Each word will be converted into a 100 lenght vector.
DIM = 100
w2v = gensim.models.Word2Vec(sentences=X, vector_size=DIM, window=10, min_count=1)
gensim.models.Word2Vec()
```

```python
# To see how many vectors are there in the 'w2v model' we do as:
len(w2v.wv.index_to_key) # for each unique word the vectors are created.
```

```python
# Let's more explore the vectors.
# To find the vectors for word 'love':
w2v.wv['love']
```

```python
w2v.wv["usa"]
```

```python
# We can also find the most similar words for a given words:
w2v.wv.most_similar("good")
```

```python
w2v.wv.most_similar("afghanistan")
```

```python
# Now here we have two methods to train the ML model: 
             # 1) we can directly use these vectors
             # 2) we can feed these vectors as initial weights to our ML model and then ML model recreate these vectors.
# So it's find that the 2nd method is better than previous one.
# So for that again we first create the tokenizer and we fit it on the X data:
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X)    
```

```python
X = tokenizer.texts_to_sequences(X) # it will not generate vectors but it will generate a set of sequence. 
```

```python
print(X[3])
```

```python
# So now what is the meaning of these sequence? For each word a particular number is assigned.
# To find that specific word for a specific number we can use tokenizer again:
# tokenizer.word_index['the']
tokenizer.word_index
```

```python
# Now we analyze the data which are in the X here. We want to know how many words are presenting each row?
plt.hist([len(x) for x in X], bins = 700)
plt.show # this will show us a histogram of total number of words present in our news.
```

* So we see that it shows that most of the words are less than 1000.

```python
# Now to see how many numbers are greater than 1000?/
nos = np.array([len(x) for x in X])
len(nos[nos>1000])
```

```python
# So next we truncate the words into a 1000.
maxlen = 1000
X = pad_sequences(X, maxlen = maxlen) # It will create every sequence as 1000 lenght.
```

```python
# Now if you check the lenght of any index of X, you will always get 1000.
len(X[1])
```

```python
len(X[1000])
```

```python
vocab_size = len(tokenizer.word_index) + 1
vocab = tokenizer.word_index
```

```python
def get_weight_matrix(model):
    weight_matrix = np.zeros((vocab_size, DIM))
    
    for word, i in vocab.items():
        weight_matrix[i] = model.wv[word]
        
    return weight_matrix
```

```python
embedding_vectors = get_weight_matrix(w2v)
```

```python
embedding_vectors.shape
```

```python
model = Sequential()           # trainable = False => says that the vectors will be not retrain.
model.add(Embedding(vocab_size, output_dim=DIM, weights = [embedding_vectors], input_length=maxlen, trainable=False))
model.add(LSTM(units=128))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer="adam", loss = 'binary_crossentropy', metrics = ['acc'])
```

```python
model.summary()
```

```python
# Next we split the dataset:
X_train, X_test, y_train, y_test = train_test_split(X, y)  
```

```python
model.fit(X_train, y_train, validation_split=0.3, epochs=3)
```

```python
y_pred = (model.predict(X_test) >= 0.5).astype(int)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
